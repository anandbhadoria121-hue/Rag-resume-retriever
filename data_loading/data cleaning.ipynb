```python
df = pd.read_csv(r"C:\Users\anand\PycharmProjects\PythonProject1\docs\extracted_resumes.csv")
df.head()
```

```python
df = pd.get_dummies(df, columns=["jobclass"], prefix="jobclass")
```

```python
df.head()
```

```python
import re

def clean_text(t):
    t = re.sub(r"[ \t]+", " ", str(t))             # collapse spaces/tabs
    t = re.sub(r"\n\s*\n+", "\n\n", str(t))        # collapse blank lines to one
    t = re.sub(r"(?<!\n)\n(?!\n)", " ", str(t))    # join single line-breaks, keep doubles
    return t.strip()

df['text'] = df['text'].apply(clean_text)
```
